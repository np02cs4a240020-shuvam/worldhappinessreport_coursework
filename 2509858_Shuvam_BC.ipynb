**Task - 3 - Outlier Detection:**

**Tasks:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[
        (df[column] < lower_bound) |
        (df[column] > upper_bound)
    ]
    return outliers

# Detect outliers for 'Score' and 'Log GDP per capita'
score_outliers = detect_outliers_iqr(df_south_asia, 'score')
gdp_outliers = detect_outliers_iqr(df_south_asia, 'Log GDP per capita')

# Identify unique outliers based on either metric
outlier_countries = pd.concat([score_outliers, gdp_outliers]).drop_duplicates(subset=['Country name'])

print("Outlier Countries in South Asia (based on Score or Log GDP per capita using 1.5*IQR):")
if not outlier_countries.empty:
    print(outlier_countries[['Country name', 'score', 'Log GDP per capita']])
else:
    print("No outliers detected based on the 1.5 * IQR rule for Score or Log GDP per capita.")

# Create a list of names for coloring in the plot
outlier_names = outlier_countries['Country name'].tolist()
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers

```python
def get_top_bottom_countries(dataframe, region_name):
    top_3 = dataframe.sort_values(by='score', ascending=False).head(3).copy()
    bottom_3 = dataframe.sort_values(by='score', ascending=True).head(3).copy()

    # Add a column for region and type for comparison
    top_3['Region'] = region_name
    top_3['Performance'] = 'Top 3'
    bottom_3['Region'] = region_name
    bottom_3['Performance'] = 'Bottom 3'
    return pd.concat([top_3, bottom_3])

sa_performers = get_top_bottom_countries(df_south_asia, 'South Asia')
me_performers = get_top_bottom_countries(df_middle_east, 'Middle East')
all_performers = pd.concat([sa_performers, me_performers])

# Print the tables
print("\nTop 3 and Bottom 3 Performers by Score:")
print(all_performers[['Region', 'Performance', 'Country name', 'score']].sort_values(by=['Region', 'Performance', 'score'], ascending=[False, False, False]))

# Plotting the comparison
plt.figure(figsize=(14, 6))
sns.barplot(
    x='Country name',
    y='score',
    hue='Region',
    data=all_performers,
    palette='Set1'
)
plt.title('Top 3 and Bottom 3 Performers by Happiness Score ')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

4. Happiness Disparity:
* Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.
* Which region has greater variability in happiness?

```python
def calculate_disparity(df):
    score_range = df['score'].max() - df['score'].min()
    score_mean = df['score'].mean()
    score_std = df['score'].std()
    # Coefficient of Variation (CV) = (Standard Deviation / Mean)
    score_cv = score_std / score_mean if score_mean != 0 else np.nan
    return score_range, score_cv

sa_range, sa_cv = calculate_disparity(df_south_asia)
me_range, me_cv = calculate_disparity(df_middle_east)

disparity_data = {
    'Region': ['South Asia', 'Middle East'],
    'Score Range (Max - Min)': [sa_range, me_range],
    'Coefficient of Variation (CV)': [sa_cv, me_cv]
}
disparity_df = pd.DataFrame(disparity_data)

print("\nHappiness Disparity (Score):")
print(disparity_df.set_index('Region').to_string(float_format='%.3f'))

greater_variability_region = "South Asia" if sa_cv > me_cv else "Middle East"
print(f"\nBased on the **Coefficient of Variation (CV)**, the **{greater_variability_region}** region has greater relative variability in happiness.")
print("The CV is preferred over the Range or Standard Deviation for comparing variability across groups with different mean values.")
```

```python
corr_metrics = ['Freedom to make life choices', 'Generosity']

# Calculate correlations for South Asia
sa_corr_freedom = df_south_asia['score'].corr(df_south_asia['Freedom to make life choices'])
sa_corr_generosity = df_south_asia['score'].corr(df_south_asia['Generosity'])
sa_corr = [sa_corr_freedom, sa_corr_generosity]

# Calculate correlations for Middle East
me_corr_freedom = df_middle_east['score'].corr(df_middle_east['Freedom to make life choices'])
me_corr_generosity = df_middle_east['score'].corr(df_middle_east['Generosity'])
me_corr = [me_corr_freedom, me_corr_generosity]

print("\nCorrelation of Score with Freedom and Generosity:")
print("South Asia Correlation:")
print(pd.Series(sa_corr, index=corr_metrics).to_string(float_format='%.3f'))
print("\nMiddle East Correlation:")
print(pd.Series(me_corr, index=corr_metrics).to_string(float_format='%.3f'))

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
plt.suptitle('Correlation of Score with Freedom and Generosity by Region', fontsize=16)

# South Asia - Freedom
sns.regplot(
    x='Freedom to make life choices',
    y='score',
    data=df_south_asia,
    ax=axes[0, 0],
    line_kws={'color':'red'}
)
axes[0, 0].set_title(f'SA: Freedom vs Score (r={sa_corr[0]:.3f}) ')
axes[0, 0].grid(alpha=0.4)

# South Asia - Generosity
sns.regplot(
    x='Generosity',
    y='score',
    data=df_south_asia,
    ax=axes[0, 1],
    line_kws={'color':'red'}
)
axes[0, 1].set_title(f'SA: Generosity vs Score (r={sa_corr[1]:.3f}) ')
axes[0, 1].grid(alpha=0.4)

# Middle East - Freedom
sns.regplot(
    x='Freedom to make life choices',
    y='score',
    data=df_middle_east,
    ax=axes[1, 0],
    line_kws={'color':'red'}
)
axes[1, 0].set_title(f'ME: Freedom vs Score (r={me_corr[0]:.3f}) ')
axes[1, 0].grid(alpha=0.4)

# Middle East - Generosity
sns.regplot(
    x='Generosity',
    y='score',
    data=df_middle_east,
    ax=axes[1, 1],
    line_kws={'color':'red'}
)
axes[1, 1].set_title(f'ME: Generosity vs Score (r={me_corr[1]:.3f}) ')
axes[1, 1].grid(alpha=0.4)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Freedom shows a clear positive link to happiness in both regions—moderate in

6. Outlier Detection:
* Identify outlier countries in both regions based on Score and GDP per Capita.
* Plot these outliers and discuss their implications.

```python
# Detect outliers for South Asia
sa_outliers_score = detect_outliers_iqr(df_south_asia, 'score')
sa_outliers_gdp = detect_outliers_iqr(df_south_asia, 'Log GDP per capita')
sa_outlier_countries = pd.concat([sa_outliers_score, sa_outliers_gdp]).drop_duplicates(subset=['Country name'])

# Detect outliers for Middle East
me_outliers_score = detect_outliers_iqr(df_middle_east, 'score')
me_outliers_gdp = detect_outliers_iqr(df_middle_east, 'Log GDP per capita')
me_outlier_countries = pd.concat([me_outliers_score, me_outliers_gdp]).drop_duplicates(subset=['Country name'])

# Combine for a single plot
combined_df = pd.concat([df_south_asia.assign(Region='South Asia'), df_middle_east.assign(Region='Middle East')])
combined_df['Is Outlier'] = False
combined_df.loc[combined_df['Country name'].isin(sa_outlier_countries['Country name']), 'Is Outlier'] = True
combined_df.loc[combined_df['Country name'].isin(me_outlier_countries['Country name']), 'Is Outlier'] = True

# Plotting the outliers
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Log GDP per capita',
    y='score',
    hue='Region',
    size='Is Outlier',
    sizes=(50, 200),
    style='Is Outlier',
    data=combined_df,
    palette='deep'
)

# Annotate outliers
for i, row in combined_df[combined_df['Is Outlier']].iterrows():
    plt.text(
        row['Log GDP per capita'] + 0.02,
        row['score'],
        row['Country name'],
        fontsize=9,
        color='black'
    )

plt.title('Outliers in South Asia and Middle East (Score vs. Log GDP per Capita) ')
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.legend(title='Region / Outlier')
plt.grid(alpha=0.4)
plt.show()

print("\nOutliers by Region (1.5 * IQR):")

print("\nSouth Asia Outliers:")
if not sa_outlier_countries.empty:
    print(sa_outlier_countries.assign(Region='South Asia')[['Country name', 'score', 'Log GDP per capita', 'Region']])
    sa_outliers_list = sa_outlier_countries['Country name'].tolist()
else:
    print("None detected.")
    sa_outliers_list = []

print("\nMiddle East Outliers:")
if not me_outlier_countries.empty:
    print(me_outlier_countries.assign(Region='Middle East')[['Country name', 'score', 'Log GDP per capita', 'Region']])
    me_outliers_list = me_outlier_countries['Country name'].tolist()
else:
    print("None detected.")
    me_outliers_list = []
```

**Discussion on Outliers and Implications:**

Afghanistan represents a clear low-end outlier in South Asia, posting very low GDP per capita and a low Happiness Score; that one country pulls regional averages down and boosts overall variability, giving the region a sense of greater instability in both economic terms and well-being metrics.
